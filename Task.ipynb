```python
def discount(dis):
    if dis>5 or dis==5:
        return(dis)
```

```python
merged_order["discount"]=merged_order.customer_id.value_counts().apply(discount).dropna()
merged_order
```

```python
Ye=merged_order.Year.unique()
Ye
```

```python
for ye in Ye:
    print(ye)
    g=merged_order.groupby("Year")
    ye_g=g.get_group(ye)
    Repeat=ye_g.customer_id.value_counts()
    Top=ye_g.product_name.value_counts()
    least=ye_g.product_name.value_counts().index[-1]
    Revanue=ye_g["total_price"].sum()
    print(f"--Repeated_order--\n{Repeat}\n--Top_selling_Sold--\n{Top}\n--Least_Sold--\n{least}\n--Revanue--\n{Revanue}")
```
